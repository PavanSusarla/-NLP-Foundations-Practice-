### Tokenization

Tokenization is the process of breaking text into smaller units such as sentences or words.


```python
 from nltk.tokenize import word_tokenize, sent_tokenize
text = "I love NLP. NLP is powerful."
docs=sent_tokenize(text)
print(docs)
```

```python
words=word_tokenize(text)
print(words)
```

### Stemming

Stemming reduces words to their base form by removing suffixes, but the output may not be a real word.


Disadvantage (VERY IMPORTANT)

❌ Produces non-meaningful words

```python
from nltk.stem import PorterStemmer,SnowballStemmer

ps=PorterStemmer()
ss=SnowballStemmer('english')

words=['eating','sleeping','history','studies']

print([ps.stem(w) for w in words])

print([ss.stem(w) for w in words])
```

### Lemmatization

Lemmatization converts words to their dictionary root form (lemma) using POS tags.

Advantage

✅ Always produces valid English words

```python
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()
for w in words:
    print(w+"------>"+lemmatizer.lemmatize(w,pos='v'))
```

### Stop Words

Stop words are commonly used words that do not add meaning to the sentence.

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
sentence="Lets go into the world of gen ai "
filtered=[w for w in word_tokenize(sentence) if w.lower() not in stop_words]
print(filtered)
```

### POS Tagging

POS tagging assigns grammatical labels like noun, verb, adjective to words.

```python
from nltk import pos_tag
import nltk

words = word_tokenize('Google is building powerful AI models')
print(pos_tag(words))
```

### Named Entity Recognition (NER)


NER identifies real-world entities like Person, Organization, Location from text.

```python
from nltk import ne_chunk

pos_tag=nltk.pos_tag(words)
print(ne_chunk(pos_tag))
```

### One Hot Encoding

One-Hot Encoding represents each word as a binary vector.

Disadvantages

❌ Sparse vectors
❌ No semantic meaning
❌ OOV problem

```python
from sklearn.preprocessing import OneHotEncoder
data=[['I'],['Love'],['NLP']]
encoder=OneHotEncoder(sparse_output=False)
print(encoder.fit_transform(data))
```

### Bag of Words (BoW)

BoW represents text based on word frequency, ignoring grammar and word order.

```python
from sklearn.feature_extraction.text import CountVectorizer

corp=['I love NLP','NLP is very powerful']

vectorizer=CountVectorizer()
X=vectorizer.fit_transform(corp)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

### TF-IDF

TF-IDF gives higher importance to rare but meaningful words across documents.

Why Better than BoW?

✅ Reduces importance of common words
✅ Captures word importance

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer()
X=tfidf.fit_transform(corp)

print(tfidf.get_feature_names_out())
print(X.toarray())
```

### Word2Vec


Word2Vec is a neural network-based embedding technique that captures semantic meaning.

Architectures

CBOW → Context → Target

Skip-gram → Target → Context

```python
#!pip install gensim
#!pip install --user gensim
from gensim.models import Word2Vec
sentences = [
    ["I", "love", "NLP"],
    ["NLP", "is", "powerful"],
    ["AI", "is", "transforming", "the", "world"],
    ["Machine", "learning", "is", "a", "part", "of", "AI"],
    ["Deep", "learning", "uses", "neural", "networks"],
    ["Neural", "networks", "learn", "patterns"],
    ["NLP", "uses", "text", "data"],
    ["Text", "data", "contains", "information"],
    ["AI", "models", "require", "training", "data"],
    ["Large", "language", "models", "are", "powerful"],
    ["Transformers", "are", "used", "in", "NLP"],
    ["BERT", "and", "GPT", "are", "transformer", "models"],
    ["GPT", "generates", "human", "like", "text"],
    ["Word2Vec", "creates", "word", "embeddings"],
    ["Embeddings", "capture", "semantic", "meaning"],
    ["AI", "engineers", "build", "intelligent", "systems"],
    ["Data", "science", "combines", "statistics", "and", "programming"],
    ["Python", "is", "popular", "for", "machine", "learning"],
    ["NLP", "applications", "include", "chatbots", "and", "translation"],
    ["Artificial", "intelligence", "is", "the", "future"]
]

model = Word2Vec(sentences, vector_size=50, window=3, min_count=1)

print(model.wv["AI"])
print(model.wv.similarity("AI", "NLP"))
```
